```python
import tkinter
from tkinter.filedialog import askopenfilename
from tkinter import END, Label, Entry, Button
from PIL import Image, ImageTk
import cv2
import os

root = tkinter.Tk()
root.title('Image Calculator')

mainHeading = Label(root, text = 'Image Calculator', font = ('Helvetica', 22))
mainHeading.grid(row = 1, column = 2, padx = 10, pady = 10)

# ------- Image A -------
# label
img1Label = Label(root, text = 'Image 1: ')
img1Label.grid(row = 2, column = 0, padx = 6, pady = 10)
# Input
imgAPath = Entry(root, width = '30')
imgAPath.grid(row = 2, column = 1, padx = 0, pady = 10)
imgAObject = ''

# ------- Image B -------
# label
img2Label = Label(root, text = 'Image 2: ')
img2Label.grid(row = 2, column = 3, padx = 10, pady = 10)
# Input
imgBPath = Entry(root, width = '30')
imgBPath.grid(row = 2, column = 4, padx = 10, pady = 10)
imgBObject = ''

# Image Load function
def ImgALoadFunc():
    # take file and returns file data
    if(imgAPath.get() == ''):
        file = askopenfilename()
        imgAPath.delete(0, END)
        imgAPath.insert(0, file)
    global imgAObject
    imgAObject = cv2.imread(imgAPath.get())[:400, :400]
    cv2.imwrite('imageA.png', imgAObject)
    showImageFunc('imageA.png', 0)
    

# Image Load Command
def ImgBLoadFunc():
    if(imgBPath.get() == ''):
        file = askopenfilename()
        imgBPath.delete(0, END)
        imgBPath.insert(0, file)
    global imgBObject
    imgBObject = cv2.imread(imgBPath.get())[:400, :400]
    cv2.imwrite('imageB.png', imgBObject)
    showImageFunc('imageB.png', 3)
    
# Image A load buttons
ImgALoadBtn = Button(root, text = 'Select Image A', command = ImgALoadFunc)
ImgALoadBtn.grid(row = 3, column = 2, padx = 10, pady = 10)

# Image B Load Buttons
ImgBLoadBtn = Button(root, text = 'Select Image B', command = ImgBLoadFunc)
ImgBLoadBtn.grid(row = 3, column = 3, padx = 10, pady = 10)


# Add image
def AddImage():
    if (canDoOperations()):
        outputImg = cv2.add(imgAObject, imgBObject)
        cv2.imwrite('add_result_img_swipe.jpg', outputImg)
        showImageFunc('add_result_img_swipe.jpg', 8)
        showSymbol(' + ')
    
# Substract image
def SubstractImage():
    if (canDoOperations()):
        outputImg = cv2.subtract(imgAObject, imgBObject)
        cv2.imwrite('sub_result_img.jpg', outputImg)
        showImageFunc('sub_result_img.jpg', 8)
        showSymbol(' - ')
    
# Multiply image
def MultiplyImage():
    if (canDoOperations()):
        outputImg = cv2.multiply(imgAObject, imgBObject)
        cv2.imwrite('multiply_result_img.jpg', outputImg)
        showImageFunc('multiply_result_img.jpg', 8)
        showSymbol(' * ')

# Divide image
def DivideImage():
    if (canDoOperations()):
        outputImg = cv2.divide(imgAObject, imgBObject)
        cv2.imwrite('divide_result_img.jpg', outputImg)
        showImageFunc('divide_result_img.jpg', 8)
        showSymbol(' / ')
    
# Buttons used for calculations
# Add button using add image
b1 = Button(root, fg = 'green', text = 'Add', command = AddImage)
b1.grid(row = 4, column = 1, padx = 10, pady = 10)

# Substract button using substract image
b2 = Button(root, fg = 'red', text = 'Substract', command = SubstractImage)
b2.grid(row = 4, column = 2, padx = 10, pady = 10)    
 
# Multiple button using add image
b3 = Button(root, fg = 'blue', text = 'Multiple', command = MultiplyImage)
b3.grid(row = 4, column = 3, padx = 10, pady = 10)   
    
# Divide button using add image
b4 = Button(root, fg = 'black', text = 'Divide', command = DivideImage)
b4.grid(row = 4, column = 4, padx = 10, pady = 10)



# Show image on window
def showImageFunc(imgPath, col):
    showImage = ImageTk.PhotoImage(Image.open(imgPath))
    label = Label(root, image = showImage)
    label.image = showImage
    label.grid(row = 10, column = col, columnspan = 3, rowspan = 4)

# Check can user can perform any action like add, sub, multiply, divide
def canDoOperations():
    if(imgAPath.get() and imgBPath.get()):
        return True
    print('Cannot perform action now')
    return False

# show operation symbol
def showSymbol(symbol):
    label = Label(root, text = symbol, fg="white", bg="blue")
    label.grid(row = 10, column = 3, pady = 10)
    # create equal symbol as well
    equallabel = Label(root, text = ' = ', fg="white", bg="blue")
    equallabel.grid(row = 10, column = 6, padx = 10, pady = 10)
    
root.mainloop()
```
